```python
import zipfile
import os

zip_path = '/content/drive/MyDrive/PlantDoc-Object-Detection/yolo.zip'
extract_path = '/content/drive/MyDrive/PlantDoc-Object-Detection/dataset'

# Unzip only if not already extracted
if not os.path.exists(extract_path):
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_path)

print("‚úÖ Dataset extracted to:", extract_path)
```

```python
os.listdir(extract_path)
```

```python
from ultralytics import YOLO

# Load a COCO-pretrained YOLOv8n model
model = YOLO("yolov8s.pt")
```

```python
model.train(
    data='/content/drive/MyDrive/PlantDoc-Object-Detection/dataset.yaml',
    epochs=30,
    imgsz=416,
    batch=8,
    name='PlantDoc-Object-Detection',
    patience=20,
    verbose=True,
    augment=True
)
```

```python
results = model.val(data="/content/drive/MyDrive/PlantDoc-Object-Detection/dataset.yaml")
```

```python
import os
import cv2
import matplotlib.pyplot as plt

# Dossier avec les r√©sultats
result_folder = "/content/runs/detect/val"

# Liste les fichiers image
images = [f for f in os.listdir(result_folder) if f.endswith(('.jpg', '.png'))]

# Affiche les 5 premi√®res images
for img_file in images[:8]:
    img_path = os.path.join(result_folder, img_file)
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.figure(figsize=(10, 10))
    plt.imshow(img)
    plt.title(img_file)
    plt.axis('off')
    plt.show()
```
